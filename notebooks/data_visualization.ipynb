```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
DATASET_PATH = "../data/product_data.csv"
df = pd.read_csv(DATASET_PATH)
df.head()
```

Visualizations will be performed on cleaned/preprocessed data

```python
df.describe()
```

```python
# Configs
sns.set(style="whitegrid", palette="Set2")
plt.rcParams['figure.figsize'] = (10, 5)
```

```python
# Price Distribution
sns.histplot(df['price'], bins=30, kde=True)
plt.title("Price Distribution of Furniture Products")
plt.xlabel("Price ($)")
plt.ylabel("Number of Products")
plt.show()
```

```python
# Top 15 Brands by Product Count
top_brands = df['brand'].value_counts().head(15)
sns.barplot(y=top_brands.index, x=top_brands.values)
plt.title("Top 15 Brands by Number of Products")
plt.xlabel("Number of Products")
plt.ylabel("Brand")
plt.show()
```

```python
# Country of Origin Distribution
country_counts = df['country_of_origin'].value_counts().head(10)
sns.barplot(x=country_counts.values, y=country_counts.index)
plt.title("Top Countries of Origin")
plt.xlabel("Number of Products")
plt.ylabel("Country")
plt.show()
```

```python
# Most Common Materials
if 'material' in df.columns:
    material_counts = df['material'].value_counts().head(10)
    sns.barplot(x=material_counts.values, y=material_counts.index)
    plt.title("Top 10 Materials Used")
    plt.xlabel("Number of Products")
    plt.ylabel("Material")
    plt.show()
```

```python
# Most Common Colors
if 'color' in df.columns:
    color_counts = df['color'].value_counts().head(10)
    sns.barplot(x=color_counts.values, y=color_counts.index)
    plt.title("Top 10 Colors")
    plt.xlabel("Number of Products")
    plt.ylabel("Color")
    plt.show()
```
